# Spotfiy Top 50 Data Analysis

## Importing packages and Data Loading

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('top50.csv', encoding = 'cp1252')
data.head()
```

```python
data.drop('Unnamed: 0', axis = 1, inplace = True)
data.head(10)
```

```python
#Statistics of DataFrame
data.describe() 
```

```python
#Checking Not Available values
data.isna().sum() 
```

```python
data.Genre
```

## Plots

```python
plt.figure(figsize = (20,10))
sns.set()
sns.countplot(data['Artist.Name'])
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (20,10))
sns.countplot(data.Genre)
plt.xticks(rotation = 90)
```

```python
#Data Cleaning
data.loc[data.Genre.str.contains('pop', case = False), 'Genre'] = 'Pop'
data.loc[data.Genre.str.contains('latin', case = False), 'Genre'] = 'Latin'
data.loc[data.Genre.str.contains('hip', case = False), 'Genre'] = 'Hip-Hop'
data.loc[data.Genre.str.contains('rap', case = False), 'Genre'] = 'Rap'
```

```python
plt.figure(figsize = (20,10))
sns.countplot(data.Genre)
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (20,15))
sns.heatmap(data.corr(), annot = True, cmap = "BuPu")
```

```python
plt.figure(figsize = (20,10))
sns.distplot(data.Popularity)
```

```python
plt.figure(figsize = (20,10))
sns.catplot(x = "Genre", y = 'Popularity', kind = "bar", data = data)
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (20,10))

sns.catplot(x = "Danceability", y = 'Energy', hue = "Genre", data = data)
plt.xticks(rotation = 90)
```

```python
sns.jointplot(x = "Popularity", y = "Energy", kind = "reg", data = data)
```
